```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
import pandas as pd

# Using Service for Chrome driver initialization
chrome_service = Service('C:/Users/stwgre/chromedriver')
driver = webdriver.Chrome(service=chrome_service)

driver.get('https://support.sas.com/en/software/all-products-support.html')

# Using the new way to find elements by CSS selector
products = driver.find_elements(By.CSS_SELECTOR, "li[style='text-align: left;']")

texts = []
documentation_links = []
learn_support_links = []

for product in products:
    # Extract the text for each product
    texts.append(product.text.split("\n")[0])
    
    # Extract the links for each product
    links = [a.get_attribute("href") for a in product.find_elements(By.TAG_NAME, "a")]
    
    # Check the number of links and append accordingly
    if len(links) > 0:
        documentation_links.append(links[0])
    else:
        documentation_links.append(None)
    
    if len(links) > 1:
        learn_support_links.append(links[1])
    else:
        learn_support_links.append(None)

for i in range(len(texts)):
    print(texts[i])
    print(documentation_links[i])
    print(learn_support_links[i])

driver.close()
```
